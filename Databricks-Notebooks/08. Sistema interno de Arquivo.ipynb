######Copiando pasta inteira
#######O parâmetro -r indica que a cópia será recursiva, ou seja, todos os arquivos e subpastas dentro da pasta serão copiados.

```python
%fs
ls '/Volumes/curso_databricks/aula/aula_volume/Bike Store/'
```

```python
%fs
cp -r '/Volumes/curso_databricks/aula/aula_volume/Bike Store/' '/Volumes/curso_databricks/aula/aula_volume/Bike Store6/' 
```

```python
%fs
ls '/Volumes/curso_databricks/aula/aula_volume/'
```

```python
dbutils.fs.cp('/Volumes/curso_databricks/aula/aula_volume/Bike Store/','/Volumes/curso_databricks/aula/aula_volume/Bike Store7/', recurse=True )
```

```python
%fs
ls '/Volumes/curso_databricks/aula/aula_volume/Bike Store7/'
```
